```python
# –ü–æ–ª—É—á–∏–º —Å—Ä–µ–∑ –ø–æ –∫–≤–∞—Ä—Ç–∏—Ä–∞–º –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ. 

# –ù–∞–π–¥–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∞ –∫–≤–∞—Ä—Ç–∏—Ä–∞ –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞. 
data_SPB = data.query('locality_name == "–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥"')
distance_max = data_SPB['cityCenters_km'].max()
print(distance_max)
```

```python
# –°–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∫–∞–∂–¥–æ–º—É –∫–∏–ª–æ–º–µ—Ç—Ä—É.

def distance_group(distance):
    for data in range(30):
        if distance <= data:
            return data

data_SPB['distance_group'] = data_SPB['cityCenters_km'].apply(distance_group)
#print(data_SPB)

# –°–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∫–∞–∂–¥–æ–º—É –∫–∏–ª–æ–º–µ—Ç—Ä—É –∏ —Å –ø–æ–º–æ—â—å—é —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∫–≤. –º–µ—Ç—Ä–∞.
data_SPB_grouped = data_SPB.pivot_table(index = 'distance_group', values = 'price_sq_meters', aggfunc = 'median')
data_SPB_grouped.columns = ['median_price_sq_meters']
data_SPB_grouped['median_price_sq_meters'] = data_SPB_grouped['median_price_sq_meters'].astype(int)
print(data_SPB_grouped) 
# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å –ø–æ–º–æ—â—å—é –≥—Ä–∞—Ñ–∏–∫–∞

data_SPB_grouped.plot(y = 'median_price_sq_meters', style='o-',figsize=(10, 5), title = 'C—Ç–æ–∏–º–æ—Å—Ç—å –æ–±—ä–µ–∫—Ç–æ–≤ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞')
plt.show()
```
